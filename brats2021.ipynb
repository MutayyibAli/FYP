```python
def load_nifty(dir, pt_id, mod):
    return nibabel.load(os.path.join(dir, pt_id + "_" + mod + ".nii.gz"))


def load_modularities(dir, pt_id):
    return [load_nifty(dir, pt_id, mod) for mod in ["flair", "t1", "t1ce", "t2"]]
```
